```python
from PIL import Image
import cv2
import numpy as np
from matplotlib import pyplot as plt
from pylab import *
import random

%matplotlib inline  

#Import the image 
image = Image.open("coast_arnat59.jpg")
img = cv2.imread('coast_arnat59.jpg',0)


print("\nVector format of the image is\n")
print(img)
edges = cv2.Canny(img,100,200)#read the edges of an image
im = image.convert('L')#Conver RGB image to GrayLevel image
print("\nEdges of the given image\n")
print(edges)
im
figure()
gray()
# show contours with origin upper left corner
contour(im, origin='image')

axis('equal')
axis('off')

figure()

#plotting the Histogram of the image


print("\nHistogram of the image is")

Gimg = color.rgb2gray(img)

hist(Gimg.flatten(), 128)
show()
```
